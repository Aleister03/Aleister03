```python
#HomeWork06
import random
def playRockPaperScissor(rounds):
  computer=("rock","paper","scissor")
  player=bot=0
  for i in range(rounds):
    user=input().lower()
    x=random.choice(computer)
    print(f'Computer : {x}')
    if user==x:
      pass
    elif user=="rock" and x=="scissor":
      player+=1
    elif x=="rock" and user=="scissor":
      bot+=1
    elif user=="scissor" and x=="paper":
      player+=1
    elif x=="scissor" and user=="paper":
      bot+=1
    elif user=="paper" and x=="rock":
      player+=1
    elif x=="paper" and user=="rock":
      bot+=1
  print(f'Your Score : {player}')
  print(f"Computer's Score : {bot}")
  if bot>player:
    print("Computer has won the game!")
  elif bot<player:
    print("You have won the game!")
  else:
    print("It's a tie!!!")
rounds=int(input())
playRockPaperScissor(rounds)
```

```python
#HomeWork06
class Store:
  def __init__(self,balance,total_items={}):
    self.balance=balance
    self.total_items=total_items
    print("New branch created!")
  def viewAllItems(self):
    if len(self.total_items)==0:
      print("There are no items in your inventory")
    else:
      print("All Items:",end="")
      out_var=""
      for items in self.total_items.keys():
        out_var+=items+", "
      print(out_var.strip(', '))
  def viewAllItemDetails(self):
    print(self.total_items)
  def add_item(self,stock):
    print('Item added:',stock[0])
    self.balance=self.balance-(stock[1]*stock[2])
    if stock[0] not in self.total_items:
      self.total_items[stock[0]]={'stock': stock[1], 'buying_price': stock[2], 'selling_price': stock[3]}
  def sell_item(self,item,stock):
    if item in self.total_items:
      if self.total_items[item]['stock']<stock:
        print(f'Sorry! {item} is not available at your desired quantity. Currently we have: {self.total_items[item]["stock"]}')
      elif self.total_items[item]['stock']>=stock:
        self.total_items[item]['stock']-=stock
        self.balance+=self.total_items[item]['selling_price']*stock
  def restock_item(self,item,stock):
    if item in self.total_items:
      self.total_items[item]['stock']+=stock
      self.balance-=self.total_items[item]['buying_price']*stock
    print(f'Restocked item: {item},Current Stock: {self.total_items[item]["stock"]}')
print("==========================")
branch1 = Store(5000)
print(f"Current Balance: {branch1.balance}")
print(f"Total items: {branch1.total_items}")
branch1.viewAllItems()
branch1.viewAllItemDetails()
print("==========================")
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["ChaCha Noodles", 10, 5, 8])
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["Sparrow Shampoo", 5, 10, 20])
print(f"Current Balance: {branch1.balance}")
print("==========================")
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.sell_item("ChaCha Noodles", 15)
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
branch1.sell_item("ChaCha Noodles", 10)
print()
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.restock_item("ChaCha Noodles", 5)
print()
branch1.viewAllItemDetails()
print()
print(f"Current Balance: {branch1.balance}\n")
print("==========================")
```

```python
#HomeWork06
class Point:
  def __init__(self, x=0, y=0):
    self.x = x
    self.y = y
    self.area = 0
  def calculate_area(self):
    return self.area
  def print_details(self):
    print("--------- Printing details ----------")
    print(f'Co-ordinate: ({self.x},{self.y})')
    print(f'Area: {self.area}')
class Circle(Point):
  def __init__(self,r,x=0,y=0):
    super().__init__(x,y)
    self.r=r
  def calculate_area(self):
    self.area=3.1416*(self.r**2)
    return self.area
  def print_details(self):
    print("--------- Printing details ----------")
    print(f'Co-ordinate: ({self.x},{self.y})')
    print(f'Area: {self.area}')
    print(f'Radius: {self.r}')
class Sphere(Circle):
  def __init__(self,r,x=0,y=0):
    super().__init__(x,y)
    self.r=r
  def calculate_area(self):
    self.area=4*3.1416*(self.r**2)
    return self.area
print("--------------1---------------")
p1 = Point(2,3)
print(f'Area of p1: {p1.calculate_area()}')
print("--------------2---------------")
p1.print_details()
print("--------------3---------------")
p2 = Point()
p2.print_details()
print("--------------4---------------")
c1 = Circle(4,0,3)
print(f'Area of c1: {c1.calculate_area()}')
print("--------------5---------------")
c1.print_details()
print("--------------6---------------")
c2 = Circle(7)
print(f'Area of c2: {c2.calculate_area()}')
print("--------------7---------------")
sph1 = Sphere(3,0,2)
print(f'Area of sph1: {sph1.calculate_area()}')
print("--------------8---------------")
sph1.print_details()
print("--------------9---------------")
sph2 = Sphere(6)
print(f'Area of sph2: {sph2.calculate_area()}')
```

```python
#HomeWork07
class Account:
  def __init__(self, account_number, balance):
    self.account_number = account_number
    self.balance = balance
    self.account_type = "General"
    self.maturity = 0
  def print_details(self):
    print("------ Account details ------")
    print(f"Account Type: {self.account_type}, Maturity: {self.maturity} years")
    print(f"Account Number: {self.account_number}, Balance: ${self.balance:.2f}")
  def deposit(self, amount):
    self.balance += amount
    print(f"Deposited ${amount:.2f}. New Balance: ${self.balance:.2f}")
  def withdraw(self, amount):
    if self.balance >= amount:
      self.balance -= amount
      print(f"Withdrew ${amount:.2f}. New Balance: ${self.balance:.2f}")
    else:
      print("Insufficient funds.")
  def year_passed(self, year):
    self.maturity += year
    print(f"Maturity of the account: {self.maturity} years")
class SavingsAccount(Account):
  def __init__(self,account_type,account_number,balance,interest_rate,limit):
    self.account_type=account_type
    self.account_number=account_number
    self.balance=balance
    self.interest_rate=interest_rate
    self.limit=limit
    self.maturity=0
  def print_details(self):
    print("------ Account details ------")
    print(f"Account Type: {self.account_type}, Maturity: {self.maturity} years")
    print(f"Account Number: {self.account_number}, Balance: ${self.balance:.2f}")
    print(f'Interest Rate: {self.interest_rate}, Minimum Limit: ${self.limit}')
  def withdraw(self,amount):
    if self.balance-amount<=self.limit:
      print("Insufficient funds.")
    else:
      self.balance -= amount
      print(f"Withdrew ${amount:.2f}. New Balance: ${self.balance:.2f}")
  def apply_interest(self):
    self.balance+=self.interest_rate*self.balance
    print(f"Interest applied. New Balance: ${self.balance}")
class FixedDepositAccount(Account):
  def __init__(self,account_type,account_number,balance,year_limit):
    self.account_type=account_type
    self.account_number=account_number
    self.balance=balance
    self.year_limit=year_limit
    self.maturity=0
  def deposit(self,amount):
    print("You can not deposit in a fixed deposit account.")
  def withdraw(self,amount):
    if self.maturity>=self.year_limit and self.balance >= amount:
      self.balance -= amount
      print(f"Withdrew ${amount:.2f}. New Balance: ${self.balance:.2f}")
    else:
      print("Can not withdraw, Account is not matured")
print("-----------1------------")
account = Account("A203", 2000)
account.print_details()
print("-----------2------------")
account.deposit(400)
account.withdraw(1500)
account.year_passed(2)
print("-----------3------------")
account.print_details()
print("-----------4------------")
savings_account = SavingsAccount("Savings","SA123", 1000, 0.05, 500)
savings_account.print_details()
print("-----------5------------")
savings_account.deposit(400)
print("-----------6------------")
savings_account.withdraw(1000)
print("-----------7------------")
savings_account.withdraw(800)
print("-----------8------------")
savings_account.apply_interest()
print("-----------9------------")
savings_account.print_details()
print("-----------10------------")
fixed_account1= FixedDepositAccount("Fixed Deposit","FDA321", 10000, 5)
fixed_account1.print_details()
print("-----------11------------")
fixed_account1.deposit(400)
print("-----------12------------")
fixed_account1.year_passed(6)
print("-----------13------------")
fixed_account1.withdraw(10000)
print("-----------14------------")
fixed_account1.print_details()
print("-----------15------------")
fixed_account2 = FixedDepositAccount("Fixed Deposit","FDA300", 50000, 7)
fixed_account2.print_details()
print("-----------16------------")
fixed_account2.withdraw(10000)
```

```python
class Rickshaw:
  def __init__(self,capacity=2):
    self.capacity=capacity
    self.passengers=[]
    self.total_fare=0
    self.distance=0
    self.booking=False
  def addPassengers(self,*user):
    if self.booking==False:
      self.distance=user[-1]
      self.booking=True
      for p in user[:-1]:
        if len(self.passengers)<self.capacity:
          self.passengers.append(p)
          print(f'{p} is ready to ride {self.distance}km(s)')
        else:
          print(f'Sorry,{p}. No more capacity')
    else:
      print(f"{len(self.passengers)} person(s) already booked the Rickshaw.\nTake another vehicle.")
  def dropPassengers(self):
    if len(self.passengers)==0:
      print("No passenger to drop.")
    else:
      self.total_fare+=self.distance*20
      self.passengers=[]
      self.booking=False
      print("All the passengers were dropped off.")
  def showInfo(self):
    out_var=f"Capacity: {self.capacity}\nCurrent Passengers: {self.passengers}\nToday's Total Fare: {self.total_fare} Taka"
    return out_var
r1 = Rickshaw()
print("1===========================")
r1.addPassengers("Rony", 3)
print("2===========================")
print(r1.showInfo())
print("3===========================")
r1.addPassengers("Shanto", "Shakib","Riyad", "Yasir", 4)
print("4===========================")
r2 = Rickshaw(3)
r2.addPassengers("Shanto", "Shakib","Riyad", "Yasir", 4)
print("5===========================")
r1.dropPassengers()
print("6===========================")
print(r2.showInfo())
print("7===========================")
r1.dropPassengers()
print("8===========================")
r1.addPassengers("Yasir", 4)
print("9===========================")
r2.dropPassengers()
r1.dropPassengers()
print("10===========================")
print(r1.showInfo())
print("11===========================")
print(r2.showInfo())
```

```python
class User:
  def __init__(self,username,email,password):
    self.username=username
    self.email=email
    self.password=password
    self.followers=set()
  def follow(self,name):
    if name not in self.followers:
      print(f"You are now following {name.username}.")
      self.followers.add(name)
    else:
      print("You are already following this user.")
  def unfollow(self,name):
    if name in self.followers:
      print(f'You have unfollowed {name.username}')
      self.followers.remove(name)
    else:
      print("You are not following this user.")
user1 = User("JohnDoe", "john.doe@gmail.com", "pass123")
user2 = User("JaneSmith", "jane.smith@gmail.com", "pass456")
user3 = User("JakePaul", "jake.paul@gmail.com", "pass789")
user1.follow(user2)
user1.follow(user2)
user3.unfollow(user2)
print('#############################')
user3.follow(user1)
user3.follow(user2)
print('#############################')
print(f"Details of the followers of {user3.username}:")
for user in user3.followers:
  print(f"UserName:{user.username} \nEmailAddress:{user.email}")
print('#############################')
user2.follow(user3)
user2.follow(user1)
user2.unfollow(user3)
user2.follow(user3)
```

```python
class Mobile:
  countryCodes = {"880": "Bangladesh", "966": "India", "455": "USA"}
  def __init__(self, model, simCardStatus):
    self.model = model
    self.__simCardStatus = simCardStatus
    print(f"Model {model} is manufactured.")
  def setSimCardStatus(self,status):
    self.__simCardStatus = status
    print("SIM card status updated successfully.")
  def getSimCardStatus(self):
    return self.__simCardStatus
  def __str__(self):
    return f"Mobile Phone Detail:\nModel:{self.model}\nSIM Card Status: {self.__simCardStatus}"
class Nokia(Mobile):
  def __init__(self,model,simCardStatus,balance=0):
    super().__init__(model,simCardStatus)
    self.model=model
    self.balance=balance
    self.dialCallHistory=[]
  def __str__(self):
    return f"Mobile Phone Detail:\nModel:{self.model}\nSIM Card Status: {self.getSimCardStatus()}\nBalance: {self.balance}Tk"
  def dialCall(self,number):
    if self.getSimCardStatus()==False:
      return "No SIM card available!"
    else:
      if self.balance==0:
        return "Insufficient balance!"
      else:
        if number[:3] not in Mobile.countryCodes:
          return "Dialing is not allowed in this region."
        else:
          self.dialCallHistory.append(number)
          return f"Dialing the number {number} to {Mobile.countryCodes[number[:3]]} region."
  def changeSIMCardStatus(self):
    if self.getSimCardStatus()==False:
      self.setSimCardStatus(True)
    else:
      self.setSimCardStatus(False)
  def rechargeSIMCard(self,amount):
    self.balance+=amount
    print(f"Recharge successful! Current balance {self.balance} TK.")
N3110 = Nokia("N3110", False)
print("#######################################")
print(N3110)
print("1======================================")
N1100 = Nokia("N1100", True,100)
print("#######################################")
print(N1100)
print("2======================================")
print(N3110.dialCall("88017196xxxx"))
print("3======================================")
N3110.changeSIMCardStatus()
print("4======================================")
print(N3110.dialCall("88017196xxxx"))
print("5======================================")
N3110.rechargeSIMCard(200)
print("6======================================")
print(N3110.dialCall("88017196xxxx"))
print("7======================================")
print(N1100.dialCall("45617196xxxx"))
print("8======================================")
print(N1100.dialCall("45517196xxxx"))
print(N1100.dialCall("96617196xxxx"))
print("9======================================")
print(f"Dial call history for {N1100.model}:{N1100.dialCallHistory}")
print(f"Dial call history for {N3110.model}:{N3110.dialCallHistory}")
```

```python
class ChefsCounter:
  reservation={}
  def __init__(self,branch_name,max_reservations=5):
    ChefsCounter.reservation[branch_name]=[]
    self.branch_name=branch_name
    self.max_reservations=max_reservations
    print(f"The {self.branch_name} branch of Chef's Counter is open for reservation!")
  def reserve(self,*customers):
    for name in customers:
      if len(ChefsCounter.reservation[self.branch_name])<self.max_reservations:
        ChefsCounter.reservation[self.branch_name].append(name)
      else:
        print(f"Sorry {customers[customers.index(name)]}, {self.max_reservations} people already made a reservation in this branch.")
  def reservation_info(self):
    var=""
    for customers in ChefsCounter.reservation[self.branch_name]:
      var+=customers+", "
    print(f'Customers who reserved in {self.branch_name} branch:\n{var.strip(", ")}')
  @classmethod
  def createNewBranch(cls,branch_name,max_reservations=5):
    return ChefsCounter(branch_name,max_reservations)
print("===============1=============")
branch1=ChefsCounter("Gulshan")
print("===============2=============")
branch1.reserve("Sam","Paul")
print("===============3=============")
branch1.reservation_info()
print("===============4=============")
branch1.reserve("John","Robin","Billy","Robert")
print("===============5=============")
branch1.reservation_info()
print("===============6=============")
branch2=ChefsCounter("Dhanmondi",7)
print("===============7=============")
branch2.reserve("Ben","Alice","Fred")
print("===============8=============")
branch2.reservation_info()
print("===============9=============")
branch2.reserve("Tom","Ken","Garet","Miles","Taylor")
print("===============10=============")
branch2.reservation_info()
print("===============11=============")
branch3=ChefsCounter.createNewBranch("100 feet")
print("===============12=============")
branch3.reserve("Harry","Bob","Jenny")
print("===============13=============")
branch3.reservation_info()
print("===============14=============")
print("Reservation Information of All Branches:",ChefsCounter.reservation)
```

```python
class Monster:
  monsterCount=0
  def __init__(self,name,power,alive=True):
    Monster.monsterCount+=1
    self.name=name
    self.power=power
    self.alive=alive
  def get_details(self):
    return f"Name:{self.name}\nPower:{self.power}\nAlive:{self.alive}"
  def attack(self,*opponents):
    if self.alive==True:
      if len(opponents)==0:
        print("No monsters to attack")
      else:
        for opponent in opponents:
          if opponent.alive==True:
            if opponent.power<self.power:
              print(f"Attack successful.{self.name} defeated {opponent.name}.")
              opponent.alive=False
              Monster.monsterCount-=1
            else:
              print(f'Attack unsuccessful. {self.name} was defeated by {opponent.name}.')
              self.alive=False
              Monster.monsterCount-=1
          else:
            print(f"Cannot attack {opponent.name}. It's not alive.")
    else:
      print(f"{self.name} is not alive to attack")
monster1 = Monster('Godzilla', 40)
monster2 = Monster('Hydra', 30)
monster3 = Monster('KingKong', 50)
print(f"Number of monsters alive:{Monster.monsterCount}")
print('1--------------------------------')
print(monster1.get_details())
print('2--------------------------------')
print(monster2.get_details())
print('3--------------------------------')
print(monster3.get_details())
print('4--------------------------------')
monster1.attack()
print('5--------------------------------')
monster1.attack(monster2)
print('6--------------------------------')
monster1.attack(monster2, monster3)
print('7--------------------------------')
print(f"Number of monsters alive:{Monster.monsterCount}")
print('8--------------------------------')
print(monster2.get_details())
print('9--------------------------------')
monster2.attack(monster1)
```

```python
class Student:
  def __init__(self,name,ID):
    self.name = name
    self.ID = ID
  def Details(self):
    return "Name: "+self.name+"\n"+"ID: "+self.ID+"\n"

class CSEStudent(Student):
  def __init__(self,name,ID,semester):
    super().__init__(name,ID)
    self.semester=semester
    self.courses={}
    self.GPA=0.0
  def Details(self):
    return super().Details()+"Current semester: "+self.semester
  def addCourseWithMarks(self,*course_marks):
    count=0
    for i in range(0,len(course_marks),2):
      if int(course_marks[count+1])>=85:
        self.courses[course_marks[count]]=4.0
      elif 80<=int(course_marks[count+1])<=84:
        self.courses[course_marks[count]]=3.3
      elif 70<=int(course_marks[count+1])<=79:
        self.courses[course_marks[count]]=3.0
      elif 65<=int(course_marks[count+1])<=69:
        self.courses[course_marks[count]]=2.3
      elif 57<=int(course_marks[count+1])<=64:
        self.courses[course_marks[count]]=2.0
      elif 55<=int(course_marks[count+1])<=56:
        self.courses[course_marks[count]]=1.3
      elif 50<=int(course_marks[count+1])<=54:
        self.courses[course_marks[count]]=1.0
      else:
        self.courses[course_marks[count]]=0.0
      count+=2
  def showGPA(self):
    self.GPA=0.0
    print(f'{self.name} has taken {len(self.courses)} courses.')
    for course,GPA in self.courses.items():
      print(f'{course}: {GPA}')
      self.GPA+=GPA
    print(f"GPA of {self.name} is {round((self.GPA*3)/(len(self.courses)*3),2)}")
Bob = CSEStudent("Bob","20301018","Fall 2020")
Carol = CSEStudent("Carol","16301814","Fall 2020")
Anny = CSEStudent("Anny","18201234","Fall 2020")
print("#########################")
print(Bob.Details())
print("#########################")
print(Carol.Details())
print("#########################")
print(Anny.Details())
print("#########################")
Bob.addCourseWithMarks("CSE111",83.5,"CSE230",73.0,"CSE260",92.5)
Carol.addCourseWithMarks("CSE470",62.5,"CSE422",69.0,"CSE460",76.5,"CSE461",87.0)
Anny.addCourseWithMarks("CSE340",45.5,"CSE321",95.0,"CSE370",91.0)
print("----------------------------")
Bob.showGPA()
print("----------------------------")
Carol.showGPA()
print("----------------------------")
Anny.showGPA()
```

```python
class Transport:
  total_traveller = 0
  def __init__(self, name, fare):
    self.name = name
    self.baseFare = fare
  def __str__(self):
    s =  "Name: "+self.name+", Base fare: "+str(self.baseFare)
    return s
class Bus(Transport):
  def __init__(self, name, fare):
    super().__init__(name, fare)
    self.passengers_count = 0
    self.passengers = {}
    print(f'Base-fare of {self.name} is {self.baseFare} Taka')
  def addPassengerWithBags(self, *passenger_details):
    for i in range(0, len(passenger_details)):
      if type(passenger_details[i]) == str:
        self.passengers[passenger_details[i]] = passenger_details[i+1]
        Transport.total_traveller += 1
        self.passengers_count += 1
  def __str__(self):
    for i in self.passengers:
      if 3<=self.passengers[i]<=5:
        passenger_fair = self.baseFare+60
        self.passengers[i] = passenger_fair
      elif self.passengers[i]>5:
        passenger_fair = self.baseFare+105
        self.passengers[i] = passenger_fair
      else:
        passenger_fair = self.baseFare
        self.passengers[i] = passenger_fair
    s = f"Name: {self.name}, Base fare: {str(self.baseFare)}\nTotal Passenger(s): {self.passengers_count}\nPassenger details:\n"
    u = "\n".join(f"Name: {key}, Fare: {self.passengers[key]}" for key in self.passengers)
    return s+u
class Train(Transport):
  def __init__(self, name, fare):
    super().__init__(name, fare)
    self.passengers_count = 0
    self.passengers = {}
    print(f'Base-fare of {self.name} is {self.baseFare} Taka')
  def addPassengerWithBags(self, *passenger_details):
    for i in range(0, len(passenger_details)):
      if type(passenger_details[i]) == str:
        self.passengers[passenger_details[i]] = passenger_details[i+1]
        Transport.total_traveller += 1
        self.passengers_count += 1
  def __str__(self):
    for i in self.passengers:
      if 3<=self.passengers[i]<=5:
        passenger_fair = self.baseFare+60
        self.passengers[i] = passenger_fair
      elif self.passengers[i]>5:
        passenger_fair = self.baseFare+105
        self.passengers[i] = passenger_fair
      else:
        passenger_fair = self.baseFare
        self.passengers[i] = passenger_fair
    s = f"Name: {self.name}, Base fare: {str(self.baseFare)}\nTotal Passenger(s): {self.passengers_count}\nPassenger details:\n"
    u = "\n".join(f"Name: {key}, Fare: {self.passengers[key]}" for key in self.passengers)
    return s+u
t1 = Bus("Volvo", 950)
print("=================================")
t1.addPassengerWithBags("David",6,"Mike",1,"Carol",3)
print("=================================")
print(t1)
print("=================================")
t2 = Train("Silk City", 850)
print("=================================")
t2.addPassengerWithBags("Bob", 2, "Simon", 4)
print("=================================")
print(t2)
print("=================================")
print("Total Passengers in Transport: ", Transport.total_traveller )
```

```python
class AppleProduct:
  def __init__(self, name, model, base_price):
    self.name = name
    self.model = model
    self.base_price = base_price
  def companyInfo(self):
    st = ("Company Name: Apple\nFouder: Steve Jobs, Steve Wozniak, Ronald Wayne\nCurrent CEO: Tim Cook\nAddress: Apple Inc, 2511 Laguna Blvd, Elk Grove, CA 95758, United States")
    return st
  def feature(self):
    st = (f"Name: {self.name}\nProduct Model: {self.model}\nHardware Quality: Excellent Hardwares\nGuarantee/ Warranty: Apple Care")
    return st
  def __str__(self):
    print('This is apple product.')
  def calculatePrice(self):
    print('Total Price:', self.base_price)
  def __add__(self, other):
    if isinstance(other, AppleProduct):
      return self.total_price + other.total_price
    return NotImplemented

class MacBookPro2020(AppleProduct):
  def __init__(self, name, model, ram, chip, tax, base_price=1299):
    super().__init__(name, model, base_price)
    self.ram = ram
    self.chip = chip
    self.tax = tax
  def feature(self):
    st = (f"Name: {self.name}\nProduct Model: {self.model}\nHardware Quality: Excellent Hardwares\nGuarantee/ Warranty: Apple Care\nRAM: {self.ram}GB\nChip: {self.chip}")
    return st
  def __str__(self):
    return f"Product Details:\n{self.feature()}\nCompany Details:\n{self.companyInfo()}"
  def calculatePrice(self):
    self.total_price = self.base_price + (self.base_price*(self.tax/100))
    print(f"Calculating Total Price:\nBase Price: {self.base_price}\nTax: {self.tax}%\nTotal Price: {self.total_price}")
    return self.total_price

class iPhone12(MacBookPro2020):
  def __init__(self, name, model, ram, chip, tax, base_price=799):
    super().__init__(name, model, ram, chip, tax, base_price)
    self.ram = ram
    self.chip = chip
    self.tax = tax
  def __str__(self):
    return super().__str__()
  def calculatePrice(self):
    self.total_price = self.base_price + (self.base_price*(self.tax/100))
    super().calculatePrice()
    return self.total_price

print('###################################')
m1 = MacBookPro2020('MacBook', 'MacBookPro2020', 8, 'M1', 10)
print(m1)
print('====================================')
m1.calculatePrice()
print('###################################')
iphone = iPhone12('iPhone', 'iPhone 12', 8, 'A14', 5)
print(iphone)
print('====================================')
iphone.calculatePrice()
print('###################################')
print('Total Price of these two products: ',end='')
print('%.2f Dollars'%(m1 + iphone))
```

```python
class Library:
  Total_book = 1000
  borrow_data = {}

  def __init__(self,n,id):
    self.student_name = n
    self.student_id = id

  def borrowbook(self):
    print("A book is borrowed!")

  def __str__(self):
    return "Library: XYZ"

class Student(Library):
  def __init__(self,n,id):
    super().__init__(n,id)
    self.book_list = []

  def borrowbook(self, book_name, unique_id = None):
    if book_name in Library.borrow_data:
      print(f"Sorry {self.student_name} ! {book_name} is already borrowed by {Library.borrow_data[book_name]}")
    else:
      Library.Total_book -= 1
      Library.borrow_data[book_name] = [self.student_name]
      super().borrowbook()
      print(f"'{book_name}' is borrowed by {self.student_name}({self.student_id})" if unique_id == None else f"'{book_name}' book with the unique id {unique_id} is borrowed by {self.student_name}({self.student_id})")
      print(f"Number of books available for borrowing = {self.Total_book}")
    self.book_list.append(book_name)

  def __str__(self):
    lst=""
    for i in self.book_list:
      lst+=i+", "
    return f"{super().__str__()}\nStudent Name: {self.student_name} ID: {self.student_id}\nBooks borrowed: {lst.strip(', ')}"

  def returnAllBooks(self):
    print(f"All Books are returned by {self.student_name}")
    for i in self.book_list:
      Library.Total_book += 1
      del Library.borrow_data[i]
    self.book_list = []

s1 = Student("Alice",18101259)
s1.borrowbook("The Alchemist", "Hdw652")
print("===============")
print(s1)
print("===============")
print(Library.borrow_data)
print("===============")
s1.borrowbook("Wuthering Heights")
print("===============")
print(s1)
print("===============")
s2= Student("David",18141777)
s2.borrowbook("The Alchemist", "Hdw652")
print("===============")
s2.borrowbook("The Vampyre")
print("===============")
print(Library.borrow_data)
print("===============")
s1.returnAllBooks()
print("===============")
print(Library.borrow_data)
```
